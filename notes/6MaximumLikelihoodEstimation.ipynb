# MLE: probability interpretation of optimization

## why we do MLE?

- **MLE is a powerful tool for building losses**

    given feature $x_i$, the larger the probability of get $y_i$, the more accurate our model is
    
    if we know $P(y=1|x)$, then we know $f^*(x)$


- **steps of learning** :  build generalized linear models from exponential families -> calculate conditional probability -> do MLE -> optimization -> do classification

## MLE and NLL for binary classification

**in general condition**

from the **MLE** perspective, we wish to **maximize** the likelihood of $\beta$. 

$p_{i}=p(x_i)$ can be any function class, later will be a neural network

because data are i.i.d, the likelihood of guess for $n$ independent samples is:

$$
L(\beta) = \prod_{i=1}^n \mathbb{P}_{\beta }(x_i,y_i)=\prod_{i=1}^n \frac{\mathbb{P}_{\beta }(y_i|x_i)}{\mathbb{P}_{\beta}(x_i)}
$$

**Assume ${\mathbb{P}_{\beta}(x_i)}$ does not depend on $\beta$**

$$
L(\beta) =\prod_{i=1}^n P_\beta(y_i|x_i)
$$

$$
NLL(\beta) =\sum_{i=1}^n [-\ln P_\beta(y_i|x_i)]
$$

$$
\hat \beta = \underset{\beta}{\arg\min}\sum_{i=1}^n [-\ln P_\beta(y_i|x_i)]
$$

for example $i$, we have guess 

$$
p_{i}=P_\beta(y_{i}| x_{i})
$$


Given the probability mass function of Bernoulli Distribution is:

$$
p_{i}=p_{i}^{y_{i}}\left ( 1- p_{i}\right )^{1-y_{i}}
$$

Then 

$$
L(p) = \prod_{i=1}^n p_{i}^{y_{i}}\left ( 1- p_{i}\right )^{1-y_{i}}
$$

Taking **natural logs** and multiplying by $-1$,

we wish to **minimize** the **negative log-likelihood** 

So let the negative log-likelihood be (total loss of n examples)

$$
NLL(p) =-\ln L(p) = \sum_{i=1}^n \left [ -y_{i}\ln(p_{i})-(1-y_{i})\ln(1-p_{i}) \right ]
$$

Then, loss of **example i** is

$$
-y_{i}\ln(p_{i})-(1-y_{i})\ln(1-p_{i})
$$

$$
\hat{p }=\underset{p \in \mathcal{F}}{\arg\min} NLL(p)=\underset{p \in \mathcal{F}}{\arg\min}\sum_{i=1}^n \left [ -y_{i}\ln(p_{i})-(1-y_{i})\ln(1-p_{i}) \right ]
$$

Empirical risk is 

$$
\hat{R}(p)=\frac{1}{n} \sum_{i=1}^n \left [ -y_{i}\ln(p_{i})-(1-y_{i})\ln(1-p_{i}) \right ]
$$
